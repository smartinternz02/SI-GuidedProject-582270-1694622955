## Machine Learning and KMeans Clustering

### Clustering for two parameters

```python
df1=df[['AnnualIncome','SpendingScore']]
df1.head()
```

```python
sns.scatterplot(x=df1["AnnualIncome"],y=df1["SpendingScore"])
```

```python
from sklearn.cluster import KMeans
errors=[]
for i in range(1,11):
  kmeans= KMeans(n_clusters=i)
  kmeans.fit(df1)
  errors.append(kmeans.inertia_)
```

```python
errors
```

```python
plt.figure(figsize=(13,6))
plt.plot(range(1,11),errors)
plt.plot(range(1,11),errors,linewidth=3,color='blue', marker='8')
plt.xlabel('No. of clusters')
plt.ylabel('errors')
plt.xticks(np.arange(1,11,1))
plt.show()
```

```python
km= KMeans(n_clusters=5)
km.fit(df1)
y=km.predict(df1)
df1['label']=y
df1.head()
```

```python
sns.scatterplot(x=df["AnnualIncome"],y=df["SpendingScore"], data=df1, hue="label", s=50, palette=['blue','red','green','orange','purple'])
```

```python
print(km.predict([[15,38]]))
```

### Clustering for all the features

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.Gender= le.fit_transform(df.Gender)
df.head()
```

```python
from sklearn import cluster
```

```python
error=[]
for i in range(1,11):
  kmeans = cluster.KMeans(n_clusters=i,init = 'k-means++',random_state=0)
  kmeans.fit(df)
  error.append(kmeans.inertia_)
```

```python
error
```

```python
plt.plot(range(1,11),error,marker='8')
plt.title('Elbow method')
plt.xlabel('number of clusters')
plt.ylabel('error')
plt.show()
```

```python
km_model = cluster.KMeans(n_clusters=6,init = 'k-means++',random_state=0)
```

```python
km_model.fit(df)
```

```python
pred = km_model.predict(df)
pred
```
