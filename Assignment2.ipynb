```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

3. Perform the Below Visualizations:


● Univariate Analysis

```python
df.corr().Price.sort_values(ascending=False)
```

```python
sns.distplot(df["living area"])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x="living area", kde=True)
plt.title('Histogram of Numeric Variable')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x="living area")
plt.title('Box Plot of Numeric Variable')
plt.show()
```

```python
sns.displot(df["living area"])
plt.show()
```

```python
sns.barplot(x=df["condition of the house"],y=df.Price)
```

● Bi - Variate Analysis


```python
sns.lineplot(x=df["living area"],y=df.Price)
```

```python
sns.lineplot(x=df["number of bedrooms"],y=df["number of floors"])
```

```python
sns.scatterplot(x=df["living area"],y=df.Price)
```

```python
sns.scatterplot(x=df["number of bedrooms"],y=df["number of floors"])
```

```python
sns.jointplot(data=df, x='number of bedrooms', y='number of bathrooms', kind='scatter')
plt.show()
```
